# 3.) Clean your data for a decision tree


```python
X = pd.get_dummies(final_df.drop('case_status',axis=1),drop_first=True)
y = final_df['case_status']
```

```python
X.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

# 4.) Fit and plot a decision tree of depth 3

```python
pruned_tree = DecisionTreeClassifier(max_depth=3)
pruned_tree.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix, accuracy_score
```

# 6.) Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
depth = []

for i in range(1,10):
    tree = DecisionTreeClassifier(max_depth=i)
    tree.fit(X_train,y_train)
    score = tree.score(X_test, y_test)
    depth.append(score)


plt.plot(range(1,10), depth)
```

The best tree seems to have a depth level of 6 based on our out of sample result.  The accuracy is about 60%
